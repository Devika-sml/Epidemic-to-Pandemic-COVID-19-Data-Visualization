```python
#Import packages

import numpy as np
import pandas as pd

# For visualization purpose
import folium
```

```python
# Retrieve LIVE COVID-19 Stats of India from Wikipedia
corona_stats = pd.read_html('https://en.wikipedia.org/wiki/COVID-19_pandemic_in_India#covid19-container',match='State/Union Territory')

#Convert to DataFrame
covid19_india_1 = pd.DataFrame(corona_stats[0])

covid19_india_1.head()
```

```python
covid19_india_1.tail()
```

```python
# cleaning COVID-19 Dataframe

# Removing unnecessary rows and columns
covid19_india = covid19_india_1.iloc[:-2,:-4]

# Renaming Attribute Names
covid19_india.columns = ['States','Total cases','Deaths','Recoveries','Active cases']

covid19_india.head()
```

```python
covid19_india.tail()
```

```python
# Retrieve latitude and longitude of states from Wikipedia
india_latlong_0 = pd.read_html('https://www.distancelatlong.com/distancecalculator/country/india',match='States')

#Convert to DataFrame
india_latlong_1 = pd.DataFrame(india_latlong_0 [0])

india_latlong_1.head()
```

```python
india_latlong_1.tail()
```

```python
covid19_india["States"].unique()
```

```python
india_latlong_1["States"].unique()
```

```python
#Replace State names to match with ccordinates data

covid19_india['States']=covid19_india.States.replace({'and Daman and Diu': "","Islands" : "",},regex=True)
covid19_india.head(8)
```

```python
#Cleaning coordinates data

india_latlong_1["States"]=india_latlong_1.States.replace({'Orissa': "Odisha","Uttaranchal" : "Uttarakhand","And " : "and "},regex=True)
india_latlong_1["States"]=india_latlong_1['States'].str.replace(r"\(.*\)","")
india_latlong_1["States"].unique()
```

```python
#Create dataframe for missing values in the coordinates data

states_latlong = pd.DataFrame({"States":["Gujarat", "Ladakh", "Telangana"], 
                    "Latitude":[22.309425, 34.209515, 17.123184 ],
                    "Longitude":[72.136230, 77.615112, 79.208824]}) 

#Append dataframe with coordinates data

india_latlong=india_latlong_1.append(states_latlong,ignore_index=True)
india_latlong.tail()
```

```python
#Remove extra spaces from States to avoid any issues while merging

for df in (india_latlong, covid19_india):
    df['States'] = df['States'].str.strip()
    
#Merging the tables covid19 and coordinates 

india_data = pd.merge(india_latlong, covid19_india, how="inner", on ='States')
india_data.tail()
```

```python
india_data["States"].unique()
```

```python
# Retrieve data from india_data table and plot it on the INDIA map
# Create the map zooming on INDIA,the lat and long of INDIA[20.5937,78.9629]

india = folium.Map(location = [20.5937,78.9629],zoom_start=5)

#add to map

for state,lat,long,total_cases,Death,Recov,Active in zip(list(india_data['States']),list(india_data['Latitude']),list(india_data['Longitude']),list(india_data['Total cases']),list(india_data['Deaths']),list(india_data['Recoveries']),list(india_data['Active cases'])):
    #create circle marker
    folium.CircleMarker(location = [lat,long],
                       radius = 5,
                       color='Red',
                       fill = True,
                       fill_color="Blue").add_to(india)
    #create marker
    folium.Marker(location = [lat,long],
                  # add information that need to be displayed on popup
                  popup=folium.Popup(('<strong><b>States  : '+state+'</strong> <br>' +
                    '<strong><b>Total Cases : '+total_cases+'</strong><br>' +
                    '<strong><font color= red>Deaths : </font>'+Death+'</strong><br>' +
                    '<strong><font color=green>Recoveries : </font>'+Recov+'</strong><br>' +
                    '<strong><font color=Orange>Active Cases : </font>'+Active+'</strong>')),max_width=200,tooltip='<strong>Click here to see Popup</strong>').add_to(india)
    
```
